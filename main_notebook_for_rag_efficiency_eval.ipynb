So we aren't using long answers because they are messy and nasty. We are going to be using only short answers, anything with long answers is added to the other pile

```python
import json
import re
import random

# Load the dataset in chunks for memory efficiency
def load_dataset(filepath):
    with open(filepath, "r") as f:
        for line in f:
            yield json.loads(line)

# Function to clean HTML-like text and extract meaningful content
def clean_text(text):
    # Remove HTML-like tags using regex
    clean = re.sub(r"<.*?>", " ", text)
    # Replace multiple whitespaces with a single space
    clean = re.sub(r"\s+", " ", clean)
    return clean.strip()

# Function to extract answer text based on tokens
def extract_answer_text(document_text, answer_annotation):
    start_token = answer_annotation.get("start_token")
    end_token = answer_annotation.get("end_token")
    if start_token is not None and end_token is not None:
        tokens = document_text.split()  # Split the document into tokens
        return " ".join(tokens[start_token:end_token])
    return None

# Separate gold and non-gold data
# Extract gold data with both long and short answers
gold_data = []
non_gold_data = []

for entry in load_dataset("./simplified_nq_dataset.jsonl"):
    document_text = entry.get("document_text", "")  # Original uncleaned text
    question_text = entry.get("question_text")
    annotations = entry.get("annotations", [])

    if annotations:
        short_answers = annotations[0].get("short_answers", [])
        # print(short_answers, len(short_answers))

        # Extract `long_answer` text if available
        #this is not needed, long answers are really messy
        # long_answer_text = None
        # if long_answer and "start_token" in long_answer and "end_token" in long_answer:
        #     long_answer_text = extract_answer_text(document_text, long_answer)

        # Extract `short_answers` text
        short_answer_texts = [
            extract_answer_text(document_text, short_answer)
            for short_answer in short_answers
        ]

        if len(short_answers) > 0:
            cleaned_text = clean_text(document_text)
            # print(short_answer_texts)
            gold_data.append({
                "question": question_text,
                "document": cleaned_text,
                "short_answers": short_answer_texts
            })
        else:
            cleaned_text = clean_text(document_text)
            print(question_text)
            non_gold_data.append({"question": question_text, "document": cleaned_text, "short_answers":None})

        if len(gold_data) >= 10000 and len(non_gold_data) >= 10000:
            break  # Stop when both gold and non-gold data is collected before random sampling

# Combine and save gold and non-gold data

#this is our question set as well, because these are the only ones that have answers
selected_gold = random.sample(gold_data, min(5000, len(gold_data)))
#either have no answers or long answers so we will only use these for additional context basically
selected_non_gold = random.sample(non_gold_data, min(5000, len(non_gold_data)))

final_data = selected_gold + selected_non_gold


# Save the combined dataset
output_file = "combined_dataset.jsonl"
with open(output_file, "w") as f:
    for item in final_data:
        f.write(json.dumps(item) + "\n")

print(f"Dataset saved to {output_file}")

```

Using the embeddings to set up chroma and add the embeddings into the document

```python
client = chromadb.Client(Settings())

# Create a new collection
collection = client.create_collection(name="simplified_nq")

# Add documents to the collection
for idx, (doc, embedding) in enumerate(zip(downsampled_docs, embeddings)):
    collection.add(
        ids=[f"doc_{idx}"],  # Unique ID for each document
        embeddings=[embedding],  # The embedding vector
        metadatas=[{"document_text": doc}]  # Metadata with the raw document text
    )

print("Documents added to the Chroma collection.")
```

Function to retrieve context given a question

```python
encoding_model = SentenceTransformer("multi-qa-MiniLM-L6-cos-v1")
def get_context(input_query, n_count):
  query_embedding = encoding_model.encode([input_query], convert_to_tensor=False)

  # Perform a search in the collection
  results = collection.query(
      query_embeddings=query_embedding,
      n_results=n_count # Number of top results to retrieve
  )
  context = ""
  for idx, metadata_list in enumerate(results.get("metadatas", [])):
    if isinstance(metadata_list, list):  # Check if the outer element is a list
        for inner_idx, metadata in enumerate(metadata_list):
            if isinstance(metadata, dict):  # Check if the inner element is a dictionary
              document_text = metadata.get("document_text", "No document text available")
              context += document_text + "\n"
  return context



```

```python
def call_server(query_with_context):
  completion = client.chat.completions.create(
    model="meta-llama/Meta-Llama-3.1-8B-Instruct",
    messages=[
      {"role": "user", "content": query_with_context}
    ]
  )
  return((completion.choices[0].message).content)
```

Code that will be used to export the ouput of our benchmark

```python
import csv

def export_benchmark_to_csv(benchmark_output, output_file="benchmark_results.csv"):
    """
    Exports benchmark results to a CSV file.

    Parameters:
        benchmark_output (list): List of tuples with (gold_index, latency, response).
        output_file (str): The name of the output CSV file.

    Returns:
        None
    """
    # Define CSV headers
    headers = ["Gold Question Index", "Latency (seconds)", "Response", "Correct Answer"]

    # Write to CSV
    with open(output_file, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerow(headers)  # Write header
        writer.writerows(benchmark_output)  # Write rows of data

    print(f"Benchmark results exported to {output_file}.")
```

```python
#output format is: the gold question index, the latency to complete, and the responsew

#question iterations is the number of times we want to run each question, i have the default output set to 5
def run_benchmark(test_set, context_count, question_iterations=5):
  benchmark_output = []
  latency_tracker = []
  for question_completed, gold_index in enumerate(test_set):
      for i in range(question_iterations):
            start = perf_counter()
            prompt = prepare_for_llama(gold_index, context_count)
            correct_answer = gold_data[gold_index]['short_answers']
            response = call_server(prompt)
            end = perf_counter()
            latency_calc = (end - start)
            benchmark_output.append((gold_index, gold_data[gold_index]['question'], latency_calc, response, correct_answer))
      print(question_completed)
  output_filename = "benchmark_results_" + "numqs_" + str(len(test_set)) + "_context_count_ " + str(context_count)  + ".csv"
  export_benchmark_to_csv(benchmark_output, output_filename)
  print(benchmark_output)
  return benchmark_output



```
